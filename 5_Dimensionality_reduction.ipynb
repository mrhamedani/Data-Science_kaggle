```python
import numpy as np
import pandas as pd
from sklearn.feature_selection import VarianceThreshold
import seaborn as sns
from sklearn.datasets import fetch_covtype
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data = fetch_covtype()
df = pd.DataFrame(data.data, columns=data.feature_names) #This dataset examines the vegetation cover in an area
df['target'] = data.target
```

```python
print(df.shape)
df.head()
```

## Considering that the table has many features, we want to find only the most important columns or features and only use them to have an easier work.

```python
x= df.drop('target', axis=1) # features
y = df['target'] # target
```

# A- Using covariance

```python
most_corr= np.abs(df.corr()['target']).sort_values(ascending=False).iloc[1:6].index.to_list()  # correlation of each feature with the target
most_corr
```

# B- Using variance

```python
varr= VarianceThreshold(threshold=1) # VarianceThreshold will remove features with variance less than 1.(That is, only features remain whose values ​​are greater than 1.)
outcome= varr.fit_transform(x)
outcome= pd.DataFrame(outcome, columns=varr.get_feature_names_out())
outcome
```

#C- using Principal Component Analysis - PCA
There is a technique that summarizes all the features in several (descending) columns

```python
scaled_data= StandardScaler().fit_transform(x) # before usinf PCA we need to scale the data
```

```python
pca= PCA(n_components=5) # we will use 5 principal components for our analysis
pca_data=pca.fit_transform(scaled_data)
df_pca = pd.DataFrame(pca_data, columns=['pca1', 'pca2', 'pca3', 'pca4', 'pca5'])

print(df_pca.shape)
df_pca.head()
```
