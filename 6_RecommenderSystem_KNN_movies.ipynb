```python
import pandas as pd
import numpy as np
from sklearn.neighbors import NearestNeighbors
```

#RecommenderSystem
Recommender systems are designed to recommend products, movies, music, articles, etc. to users based on past interests and behaviors.
1- Collaborative Filtering:
Based on the user's behavior (like, search, etc.), it suggests content that people similar to that person follow

2- Content-Based Filtering:Based on the user's behavior (like and search, etc.), it gives content similar to what we were looking for


```python
movies= pd.read_csv('movies.csv')
movie_titles = movies[["movieId", "title"]]
movies
```

```python
#Convert genres to binary columns
df_genres = movies["genres"].str.get_dummies(sep="|")   #One hot encoding for genres column

# Adding new columns to the main dataframe
movies = pd.concat([movies, df_genres], axis=1)
```

```python
movies.drop(columns=["genres"], inplace=True)
movies
```

#K-Nearest Neighbors (KNN)
KNN is a similarity-based algorithm used to find the closest examples in a feature space. In recommender systems, KNN is used to find similar movies based on their characteristics (such as genres).
this model is lazy

**K is the number of nearest neighbors**

```python
knn = NearestNeighbors(metric="cosine", algorithm="brute")
knn.fit(movies.drop(columns=["movieId", "title"])) # train model
```

```python
def recommend_movies(movie_name, n_recommendations=5):
  # Finding the movie index
    movie_index = movies[movies["title"] == movie_name].index[0]
    # Finding the movie recommendations
    _, indices = knn.kneighbors([movies.iloc[movie_index, 2:]], n_neighbors=n_recommendations+1)
    recommended_indices = indices.flatten()[1:]  # Remove the movie itself from the recommendations
    print(movie_name)
    for idx in recommended_indices:
      print(f"âœ… {movie_titles.iloc[idx]['title']}")
```

```python
recommend_movies(movie_name="Toy Story (1995)")
```
